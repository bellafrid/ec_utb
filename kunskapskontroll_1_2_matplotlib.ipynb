```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

OO-style & Pyplot, OO-style is prefered when working on larger projects, or if functions and scripts are 
intended to be reused. Pyplot is on the other hand more convenient for quick or small projects.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: The whole part of a figure, can contains everything below

Axes: Axes is needed to create the visual data, like a frame.

Axis: Sets the scales and limits of axes.

Artist: Pretty much everything that's visible, figures, colors, sizes etc.

### When plotting in Matplotlib, what is the expected input data type?

Arrays or anything that can convert to an array - before plotting that is.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_title("y = x^2")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1,2, layout='constrained')
fig.suptitle('Bellas plots', size = 20)

axs[0].scatter(random_data_x, random_data_y, color = "silver", edgecolor = "black")
axs[0].set_title("Random scatter")

axs[1].bar(names, values, color = "gold", edgecolor = "black")
axs[1].set_title("Fruit bar")
```
